## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'АВ222С96'`

Результат: 
`Номер АВ222С валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re

# пример почему-то не соответствует заданию: АВ222С96 - не валидный, т.к. 2 буквы в начале и 1 в середине
# валидным будет - А222ВС96

def get_region(number):
    result = re.findall(r"^([АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2})(\d{2,3})$", number)
    if len(result) != 0 and len(result[0]) == 2:
        return result[0][1]

car_id = input("Введите номер: ")
region = get_region(car_id)
if region is not None:
    print(f"Результат: Номер {car_id} валиден. Регион: {region}")
else:
    print(f"Результат: Номер {car_id} не валиден")
````

## Задание 2

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:  

`some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'`

Результат: 
`Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.`

````python
import re

def remove_repeats(st):
    return re.sub(r"(\w+)(?:\W+)(?=\1\W+|\1$)+", "", st)


some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'
result_string = remove_repeats(some_string)
print(result_string)
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
import re

def get_acro(st):
    return re.sub(r"(?<=\w)(\w+|\s+)", "", st).upper()

some_words = input("Введите строку: ")
result_string = get_acro(some_words)
print(result_string)
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', xyz@gmail.com]`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
import re

def get_domains_count(lst):
    result = {}
    for address in lst:
        domains = re.findall(r"(?:\w+)@([a-zA-Z0-9.]+)", address)
        for domain in domains:
            result[domain] = result.get(domain, 0) + 1
    return result


# как-то странно, последний элемент списка в задании вообще без кавычек

emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', 'xyz@gmail.com']

result = get_domains_count(emails)
if len(result) == 0:
    print("Нет данных")
else:
    for domain, count in result.items():
        print(f"{domain}: {count}")
````

## Задание 5 (необязательное)

Напишите функцию, которая будет подсчитывать сколько слов начинается на гласные, а сколько на согласные буквы в тексте (текст может быть написан как с использованием букв кириллицы, так и латиницы).

Пример работы программы:  

`some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'`

Результат: 
```
Слов на гласные буквы: 9
Слов на согласные буквы: 21 
```

````python
import re

def get_acro(st):
    fst = re.sub(r"(?<=\w)(\w*\W+)", "", st)
    
    # Неожиданно: Y - может быть гласной и согласной,
    # но в начале слова она считается согласной.
    # Об этом есть в Википедии, а вот тут с примерами:
    # http://sly2m.com/blog/2010/09/21/bukva-y-glasnaya-ili-soglasnaya
    
    snd = re.findall(r"[аеёиоуыэюяaeiou]", fst, flags=re.IGNORECASE)
    return {"vowels": len(snd), "consonants": len(fst) - len(snd)}


some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'
result = get_acro(some_text)
print(f"Слов на гласные буквы: {result['vowels']}")
print(f"Слов на согласные буквы: {result['consonants']}")
````

## Задание 6 (необязательное)
Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:   
+7-xxx-xxx-xx-xx  
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.
Примеры работы программы:  

1. `phone = '+7 955 555-55-55'`

Результат: 
`+7-950-555-55-55`

2. `phone = '8(955)555-55-55'`

Результат: 
`+7-950-555-55-55`

3. `phone = '+7 955 555 55 55'`

Результат: 
`+7-950-555-55-55`

4. `phone = '7(955) 555-55-55'`

Результат: 
`+7-950-555-55-55`

5. `phone = '423-555-55-5555'`

Результат: 
`Номер не валиден`

6. `phone = '123-456-789'`

Результат: 
`Номер не валиден`

````python
import re

# Не совсем ясно форматы для какой страны нужны,
# если брать не только Россию, то задача становится слишком объемной, см.:
# https://ru.wikipedia.org/wiki/E.164
# https://ru.wikipedia.org/wiki/%D0%A2%D0%B5%D0%BB%D0%B5%D1%84%D0%BE%D0%BD%D0%BD%D1%8B%D0%B5_%D0%BA%D0%BE%D0%B4%D1%8B_%D1%81%D1%82%D1%80%D0%B0%D0%BD
# https://habr.com/ru/post/279751/

def get_number_v1(phone):
    '''
    Строгая проверка соотвествия формату
    '''
    s = re.findall(r"^(?:\s*\+?[78]\s*[(-]?\s*)(\d{3})(?:\s*[)-]?\s*)(\d{3})(?:\s*[-]?\s*)(\d{2})(?:\s*[-]?\s*)(\d{2})(?:\s*)$", phone)
    if len(s) != 0:
        return f"+7-{s[0][0]}-{s[0][1]}-{s[0][2]}-{s[0][3]}"

def get_number_v2(phone):
    '''
    Менее строгая проверка: поддерживаются различные способы группировки типа: 8-955-55-55-555
    Однако, также считаются валидными нежелательные в некоторых случаях номера типа: 8-955-(55-55-555
    '''
    pre = re.sub(r"[-() ]", "", phone)
    s = re.findall(r"^(?:\+?[78])(\d{3})(\d{3})(\d{2})(\d{2})$", pre)
    if len(s) != 0:
        return f"+7-{s[0][0]}-{s[0][1]}-{s[0][2]}-{s[0][3]}"


phones = [
          # валидные
          '+7 955 555-55-55', '8(955)555-55-55', '+7 955 555 55 55', '7(955) 555-55-55',
          '8 (955)555-55-55', '7 (955) 555-55-55', '89555555555', '7 (955) 5555555',
          '8 (955) 555- 55 -55', '7 (955) 555  55  55', '  89555555555 ', '7 (955) 555 - 55 - 55',
          '+7-955-555-55-55',
    
          # валидные или невалидные в зависимости от алгоритма
          '8-955-55-55-555', '8-955-(55-55-555',
    
          # невалидные
          '423-555-55-5555', '123-456-789'
]

for phone in phones:
    result = get_number_v1(phone)
    #result = get_number_v2(phone)
    if result is None:
        print(f"Номер не валиден")
    else:
        print(f"Номер валиден: {result}")
````
