**Задание**

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)

Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?

```
777: 200 руб.  
999: 100 руб.  
555: 50 руб.  
333: 15 руб.  
111: 10 руб.  
*77: 5 руб.  
**7: 3 руб.  
*00: 2 руб.  
**0: 1 руб.  
```
\* - любая цифра

````python
import numpy as np

def check_win(val):
    if val == 777:
        return 200
    elif val == 999:
        return 100
    elif val == 555:
        return 50
    elif val == 333:
        return 15
    elif val == 111:
        return 10
    elif val % 100 == 77:
        return 5
    elif val % 10 == 7:
        return 3
    elif val % 100 == 0:
        return 2
    elif val % 10 == 0:
        return 1
    else:
        return 0

test_count = 10000000
tests = np.random.randint(0, 1000, size=test_count) # можно было бы использовать np.random.uniform(), но
                                                    # np.random.randint() также генерирует с равном.
                                                    # распределением, при этом значения сразу целые
summa = -test_count # заплатим заранее за все попытки
for test in tests:
    summa += check_win(test)

sr = summa / test_count
print(f"Игроку игра не выгодна, прибыль в среднем за каждую игру составляет {sr} рублей.")
````

**Дополнительное задание повышенной сложности**

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

$P(777)=P(999)=P(555)=P(333)=P(111)=\frac{1}{1000}$ - вероятность выпадения 777, 999, 555, 333, 111 по отдельности.   
$P(*77)=\frac{9}{1000}$ - вероятность выпадения \*77, за исключением 777.  
$P(**7)=\frac{90}{1000}$ - вероятность выпадения \*\*7, за исключением 777 и \*77.  
$P(*00)=\frac{10}{1000}$ - вероятность выпадения \*00.  
$P(**0)=\frac{90}{1000}$ - вероятность выпадения \*\*0, за исключением \*00.  
Выйгрыш за игру $W$ найдем как произведение вероятностей на соответствующий выйгрыш:  
$W=\frac{1\cdot200 + 1\cdot100 + 1\cdot50 + 1\cdot15 + 1\cdot10 + 9\cdot5 + 90\cdot3 + 10\cdot2 + 90\cdot1}{1000}=\frac{800}{1000}=0,8$ рублей.  
При этом на каждую игру он тратит $B=1$ рубль, т.е. его прибыль $S$ составит:  
$S=W-B=0,8 - 1 = -0,2$ рублей.  
Полученное в ходе моделирования значение близко к расчетному. И чем больше игр, тем оно получается ближе.
