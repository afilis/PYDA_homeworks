**Задание 1**  
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')

rating = 5.0
rated = ratings[ ratings['rating'] == rating ]
counts = rated['movieId'].value_counts()
if len(counts):
    mvid = counts.idxmax()
    name = movies[ movies['movieId'] == mvid ]['title']
    if len(name):
        print(f"Фильм \"{name.iloc[0]}\" набрал больше всего оценок {rating} ({counts.max()})")
    else:
        print(f"Фильм с ID = {mvid} набрал больше всего оценок {rating} ({counts.max()})")
else:
    print(f"Нет фильмов с оценкой {rating}")
```

**Задание 2**  
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')

power_balt = power[power['country'].isin(["Estonia", "Latvia", "Lithuania"]) &
                   power['category'].isin([4, 12, 21]) &
                   power['year'].isin([x for x in range(2005, 2011)])
                  ]['quantity'].sum()
print("Суммарное потребление:", power_balt)
```
